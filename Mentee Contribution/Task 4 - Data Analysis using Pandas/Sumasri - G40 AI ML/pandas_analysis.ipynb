```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df = pd.read_csv("dataset.csv", encoding="latin1")
```

```python
print(df.shape)
df.head()
```

```python
# First 10 rows
print("FIRST 10 ROWS")
display(df.head(10))

# Last 10 rows
print("LAST 10 ROWS")
display(df.tail(10))

# Shape
print("\nShape:")
print(df.shape)

# Column names
print("\nColumns:")
print(df.columns.tolist())

# Data types
print("\nData Types:")
print(df.dtypes)

# Complete information
print("\nDataset Info:")
df.info()
```

```python
# Missing values
print("Missing Values:")
print(df.isnull().sum())

# Missing value percentages
print("\nMissing Percentage:")
print((df.isnull().sum() / len(df)) * 100)

# Duplicate rows
print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
# Standardize column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

# Remove duplicate rows
df = df.drop_duplicates()

print("New Shape:", df.shape)
```

```python
df.head()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])

print(df[["order_date", "ship_date"]].dtypes)
```

```python
# Summary statistics
display(df.describe())

# Categorical summary
display(df.describe(include="object"))
```

```python
print(df["category"].value_counts())

print(df["region"].value_counts())

print(df["segment"].value_counts())
```

```python
# 1. High sales orders
high_sales = df[df["sales"] > 1000]
print("High Sales Orders:", high_sales.shape)

# 2. Technology products
technology = df[df["category"] == "Technology"]
print("Technology Orders:", technology.shape)

# 3. West region orders
west_region = df[df["region"] == "West"]
print("West Region Orders:", west_region.shape)

# 4. Profitable orders
profitable = df[df["profit"] > 0]
print("Profitable Orders:", profitable.shape)

# 5. Large quantity orders
large_quantity = df[df["quantity"] >= 5]
print("Large Quantity Orders:", large_quantity.shape)
```

```python
# Top 10 sales
top_sales = df.sort_values(by="sales", ascending=False)
display(top_sales.head(10))

# Top 10 profits
top_profit = df.sort_values(by="profit", ascending=False)
display(top_profit.head(10))
```

```python
category_summary = df.groupby("category").agg(
    total_sales=("sales", "sum"),
    average_sales=("sales", "mean"),
    total_profit=("profit", "sum"),
    total_quantity=("quantity", "sum")
)

category_summary
```
