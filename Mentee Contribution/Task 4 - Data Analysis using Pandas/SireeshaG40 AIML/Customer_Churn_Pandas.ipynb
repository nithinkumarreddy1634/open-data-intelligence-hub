Cell 21: GroupBy (Single-Level)

```python
category_summary = df.groupby("store").agg(
    Record_Count=("weekly_sales", "count"),
    Total_Sales=("weekly_sales", "sum"),
    Average_Sales=("weekly_sales", "mean"),
    Minimum_Sales=("weekly_sales", "min"),
    Maximum_Sales=("weekly_sales", "max")
).reset_index()

category_summary
```

Cell 22: Multi-Level GroupBy

```python
region_category_summary = df.groupby(["store", "holiday_flag"]).agg(
    Record_Count=("weekly_sales", "count"),
    Total_Sales=("weekly_sales", "sum"),
    Average_Sales=("weekly_sales", "mean")
).reset_index()

region_category_summary
```

Cell 23: Top 10 Stores

```python
top10 = category_summary.sort_values(
    by="Total_Sales",
    ascending=False
).head(10)

top10
```

Cell 24: Pivot Table

```python
pivot = pd.pivot_table(
    df,
    values="weekly_sales",
    index="store",
    columns="holiday_flag",
    aggfunc="sum",
    fill_value=0
)

pivot
```

Cell 25: Feature Engineering

```python
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day_name"] = df["date"].dt.day_name()

df.head()
```

Cell 26: Sales Category

```python
def sales_category(value):
    if value >= 2000000:
        return "High"
    elif value >= 1000000:
        return "Medium"
    else:
        return "Low"

df["sales_category"] = df["weekly_sales"].apply(sales_category)

df.head()
```

Cell 29: Monthly Sales Trend

```python
monthly_sales = df.groupby("month").agg(
    Total_Sales=("weekly_sales", "sum")
).reset_index()

plt.figure(figsize=(10,5))

plt.plot(
    monthly_sales["month"],
    monthly_sales["Total_Sales"],
    marker="o"
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.show()
```

Cell 33: Outlier Detection (IQR)

```python
Q1 = df["weekly_sales"].quantile(0.25)

Q3 = df["weekly_sales"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR

upper = Q3 + 1.5 * IQR

outliers = df[
    (df["weekly_sales"] < lower) |
    (df["weekly_sales"] > upper)
]

outliers.head()
```
