```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
products = pd.DataFrame({
    "Product": [
        "Laptop",
        "Gaming Laptop",
        "Laptop Bag",
        "Wireless Mouse",
        "Keyboard",
        "Gaming Keyboard",
        "Monitor",
        "Printer",
        "Smartphone",
        "Smart Watch",
        "Bluetooth Speaker",
        "Headphones",
        "Tablet",
        "USB Cable",
        "Power Bank"
    ]
})

products
```

```python
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(products['Product'])

print(tfidf_matrix.shape)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix)

print(cosine_sim)
```

```python
indices = pd.Series(products.index, index=products['Product'])

indices
```

```python
def recommend(product_name):

    if product_name not in indices:
        return "Product not found."

    idx = indices[product_name]

    similarity_scores = list(enumerate(cosine_sim[idx]))

    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    similarity_scores = similarity_scores[1:6]

    product_indices = [i[0] for i in similarity_scores]

    return products.iloc[product_indices]
```

```python
print("Recommended Products for Laptop")
recommend("Laptop")
```

```python
print("Recommended Products for Keyboard")
recommend("Keyboard")
```

```python
print("Recommended Products for Smartphone")
recommend("Smartphone")
```
