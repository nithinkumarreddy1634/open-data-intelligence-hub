## AI/ML Interview Questions: Regression

#### 1. What is regression in machine learning, and when is it used?

Regression is a supervised learning technique used to predict continuous numerical values such as salary, house price, or temperature.

#### 2. What is the difference between simple linear regression and multiple linear regression?

Simple Linear Regression: Uses one independent variable.
Multiple Linear Regression: Uses two or more independent variables.

#### 3. What assumptions are made by linear regression?

-Linear relationship between variables

-No multicollinearity

-Independence of errors

-Constant variance (homoscedasticity)

-Errors are normally distributed

#### 4. What is the difference between a dependent variable and an independent variable?

-Dependent Variable: Output/target variable.

-Independent Variable: Input/feature used for prediction.

#### 5. What do the slope and intercept represent in a linear regression model?

-Slope (m): Change in output for a one-unit change in input.

-Intercept (c): Value of output when input is zero.

#### 6. What is the purpose of the cost function in regression?

The cost function measures the error between actual and predicted values. The goal is to minimize this error.

#### 7. What is Mean Squared Error (MSE), and why is it commonly used?

MSE is the average of squared differences between actual and predicted values. It heavily penalizes large errors.

#### 8. Difference between MAE, MSE, and RMSE?

| Metric | Meaning                |
| ------ | ---------------------- |
| MAE    | Average absolute error |
| MSE    | Average squared error  |
| RMSE   | Square root of MSE     |


#### 9. What is the R-squared score?

R² measures how much variation in the target variable is explained by the model.

Range: 0 to 1

Higher value = Better model.

#### 10. What is adjusted R-squared, and why is it useful?

Adjusted R² considers the number of features and prevents misleading improvement when unnecessary features are added.

#### 11. What is overfitting in a regression model?

Overfitting occurs when a model learns training data too well and performs poorly on new data.

#### 12. What is underfitting in a regression model?

Underfitting occurs when a model is too simple and cannot capture patterns in the data.

#### 13. What is multicollinearity, and how does it affect regression?

Multicollinearity occurs when independent variables are highly correlated. It makes coefficient estimates unstable.

#### 15. What is regularization, and why is it used?

Regularization adds a penalty to the model to reduce overfitting and improve generalization.

#### 16. Difference between Ridge, Lasso, and Elastic Net Regression?

| Method      | Penalty                  |
| ----------- | ------------------------ |
| Ridge       | L2 Regularization        |
| Lasso       | L1 Regularization        |
| Elastic Net | Combination of L1 and L2 |


#### 17. How does Lasso regression perform feature selection?

Lasso reduces some coefficients to exactly zero, automatically removing less important features.

#### 18. What is polynomial regression?

Polynomial regression models non-linear relationships by adding polynomial terms such as x², x³, etc.

#### 19. How do outliers affect a regression model?

Outliers can significantly change the regression line and reduce model accuracy.

#### 20. How would you evaluate the performance of a regression model?

Common evaluation metrics:

MAE

MSE

RMSE

R² Score

Adjusted R²
