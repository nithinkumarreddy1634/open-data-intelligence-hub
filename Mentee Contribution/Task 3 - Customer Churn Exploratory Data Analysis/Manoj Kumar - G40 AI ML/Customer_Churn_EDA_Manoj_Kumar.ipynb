```python
import os
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# Dynamically search and find the file in the workspace
file_name = None
for root, dirs, files in os.walk("."):
    for f in files:
        if "Customer_Churn_Dataset_Manoj" in f:
            file_name = os.path.join(root, f)
            break

if file_name is None:
    for root, dirs, files in os.walk("/content"):
        for f in files:
            if "Customer_Churn_Dataset_Manoj" in f:
                file_name = os.path.join(root, f)
                break

if file_name:
    print(f"Found and loading file from: {file_name}")
    df = pd.read_csv(file_name)

    print("\n=== DATASET OVERVIEW ===")
    print(f"Total Rows: {df.shape[0]}")
    print(f"Total Columns: {df.shape[1]}")
    print("\n=== COLUMN DETAILS & DATA TYPES ===")
    print(df.info())

    df.head()
else:
    print("Error: Could not locate the file. Please re-upload it to the sidebar.")
```

```python
df.columns = df.columns.str.strip()

if 'TotalCharges' in df.columns:
    df['TotalCharges'] = df['TotalCharges'].replace(" ", np.nan)
    df['TotalCharges'] = df['TotalCharges'].astype(float)

df = df.dropna()

if 'customerID' in df.columns:
    df = df.drop(columns=['customerID'])

df = df.drop_duplicates()

print("=== CLEANED DATASET OVERVIEW ===")
print(f"Total Cleaned Rows: {df.shape[0]}")
print(f"Total Cleaned Columns: {df.shape[1]}")
```

```python
target_column = None
for col in df.columns:
    if "churn" in col.lower():
        target_column = col
        break

if target_column:
    print(f"Target column found: {target_column}")

    fig1 = px.histogram(df, x=target_column, title="Overall Customer Churn Count",
                        color=target_column, color_discrete_sequence=["#2ec4b6", "#e71d36"])
    fig1.show()

    fig2 = px.pie(df, names=target_column, title="Customer Churn Percentage Ratio",
                  hole=0.4, color_discrete_sequence=["#2ec4b6", "#e71d36"])
    fig2.show()
else:
    print("Available columns in your dataset:")
    print(list(df.columns))
```

```python
tenure_col = next((c for c in df.columns if "tenure" in c.lower()), None)
monthly_col = next((c for c in df.columns if "monthly" in c.lower()), None)

if tenure_col and target_column:
    fig3 = px.box(df, x=target_column, y=tenure_col, title=f"{tenure_col} Distribution vs Churn Status", color=target_column)
    fig3.show()

if monthly_col and target_column:
    fig4 = px.histogram(df, x=monthly_col, color=target_column, title=f"{monthly_col} Distribution by Churn", barmode="overlay")
    fig4.show()
```

```python
contract_col = next((c for c in df.columns if "contract" in c.lower()), None)
payment_col = next((c for c in df.columns if "payment" in c.lower()), None)

if contract_col and target_column:
    fig5 = px.histogram(df, x=contract_col, color=target_column, barmode="group", title="Churn Variance across Contract Terms")
    fig5.show()

if payment_col and target_column:
    fig6 = px.histogram(df, x=payment_col, color=target_column, barnorm="percent", title="Impact of Payment Mechanisms on Attrition")
    fig6.show()
```

```python
total_charges_col = next((c for c in df.columns if "total" in c.lower() and "charge" in c.lower()), None)

if tenure_col and total_charges_col and target_column:
    fig7 = px.scatter(df, x=tenure_col, y=total_charges_col, color=target_column, title="Tenure vs Total Charges Interplay")
    fig7.show()

numeric_cols = df.select_dtypes(include=[np.number])
corr_matrix = numeric_cols.corr()
fig8 = px.imshow(corr_matrix, text_auto=True, title="Numerical Features Correlation Heatmap", color_continuous_scale="Viridis")
fig8.show()
```

```python
df.to_csv("Cleaned_Customer_Churn_Dataset_ManojKumar.csv", index=False)
print("Cleaned dataset successfully exported!")
```
