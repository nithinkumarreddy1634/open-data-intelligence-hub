# Customer Churn Exploratory Data Analysis

This notebook performs a complete exploratory analysis of customer churn and prepares a simple predictive model.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Load the dataset
df = pd.read_csv("subscription_customer_churn.csv")

# Convert numeric columns
for col in ["tenure", "monthly_charges", "total_charges"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

print("=" * 50)
print("First 5 Rows")
print("=" * 50)
print(df.head())

print("\n" + "=" * 50)
print("Dataset Shape")
print("=" * 50)
print(df.shape)

print("\n" + "=" * 50)
print("Dataset Information")
print("=" * 50)
print(df.info())

print("\n" + "=" * 50)
print("Summary Statistics")
print("=" * 50)
print(df.describe())

print("\n" + "=" * 50)
print("Missing Values")
print("=" * 50)
print(df.isnull().sum())

print("\n" + "=" * 50)
print("Duplicate Rows")
print("=" * 50)
print(df.duplicated().sum())

# Churn distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="churn", data=df)
plt.title("Customer Churn Distribution")
plt.show()

# Gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="gender", data=df)
plt.title("Gender Distribution")
plt.show()

# Contract type distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="contract_type", data=df)
plt.title("Contract Type Distribution")
plt.xticks(rotation=20)
plt.show()

# Payment method distribution
plt.figure(figsize=(7, 4))
sns.countplot(x="payment_method", data=df)
plt.title("Payment Method Distribution")
plt.xticks(rotation=20)
plt.show()

# Monthly charges distribution
plt.figure(figsize=(6, 4))
plt.hist(df["monthly_charges"], bins=6)
plt.title("Monthly Charges Distribution")
plt.xlabel("Monthly Charges")
plt.ylabel("Customers")
plt.show()

# Tenure distribution
plt.figure(figsize=(6, 4))
plt.hist(df["tenure"], bins=6)
plt.title("Tenure Distribution")
plt.xlabel("Tenure (Months)")
plt.ylabel("Customers")
plt.show()

# Monthly charges vs churn
plt.figure(figsize=(6, 4))
sns.boxplot(x="churn", y="monthly_charges", data=df)
plt.title("Monthly Charges vs Churn")
plt.show()

# Tenure vs churn
plt.figure(figsize=(6, 4))
sns.boxplot(x="churn", y="tenure", data=df)
plt.title("Tenure vs Churn")
plt.show()

# Correlation heatmap
numeric_df = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Save cleaned dataset
df.to_csv("cleaned_subscription_customer_churn.csv", index=False)

print("\n" + "=" * 50)
print("EDA Completed Successfully")
print("=" * 50)

print("""
Key Observations:
1. Dataset loaded successfully.
2. Missing values and duplicate rows were checked.
3. Customer churn distribution was analyzed.
4. Contract type and payment method were visualized.
5. Monthly charges and tenure distributions were explored.
6. Boxplots compared monthly charges and tenure with churn.
7. Correlation heatmap showed relationships among numerical features.
8. Cleaned dataset was saved as 'cleaned_subscription_customer_churn.csv'.
""")
```
