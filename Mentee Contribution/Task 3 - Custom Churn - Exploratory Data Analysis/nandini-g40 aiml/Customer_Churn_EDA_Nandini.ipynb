## Outlier Detection

Outliers are extreme values that differ significantly from other data points. We detect them using boxplots and the IQR (Interquartile Range) method.

1. Statistical Method 

```python
Q1 = df["MonthlyCharges"].quantile(0.25)
Q3 = df["MonthlyCharges"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["MonthlyCharges"] < lower_bound) | (df["MonthlyCharges"] > upper_bound)]

print("Number of outliers:", outliers.shape[0])
```

## Univariate Analysis
Univariate analysis is the process of analyzing a single variable at a time to understand its distribution, patterns, and characteristics. It helps in summarizing data using statistical measures like mean, median, and frequency, and is often visualized using charts such as histograms and count plots.

1. Churn count plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Churn', data=df)
plt.title("Churn Count Plot")
plt.show()
```

### Observation:
The churn count plot shows that more customers have not churned compared to those who have churned. However, a noticeable portion of customers are still leaving the service.

### Business Meaning:
This indicates that while the company retains most customers, churn is still a significant issue. Improving customer satisfaction and retention strategies can further reduce churn.


2. Churn Percentage

```python
df['Churn'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Churn Percentage")
plt.ylabel("")
plt.show()
```

### Observation:
The pie chart shows the percentage distribution of churned and non-churned customers. Non-churned customers form the majority, while churned customers make up a smaller but important portion.

### Business Meaning:
Even a smaller churn percentage can lead to revenue loss. The company must focus on reducing this churn rate to improve long-term profitability.

3. Monthly Charges


```python
plt.hist(df['MonthlyCharges'], bins=20)
plt.title("Monthly Charges Distribution")
plt.xlabel("Monthly Charges")
plt.ylabel("Frequency")
plt.show()
```

### Observation:
The histogram shows that most customers fall within a certain range of monthly charges, with fewer customers at very high charge levels.

### Business Meaning:
Pricing is an important factor in customer distribution. Extremely high charges may affect affordability and could lead to churn if not managed properly.


4.  Outliners in Tenure Distribution 


Tenure distribution shows how long customers have stayed with the company, measured in months or years. It is a univariate analysis because we are analyzing only one variable: tenure, without comparing it with any other feature.

```python
sns.boxplot(y=df['tenure'])
plt.title("Outliers in Tenure")
plt.show()
```

### Observation:
The boxplot shows that most customers have moderate tenure, and there are no extreme outliers in the data.

### Business Meaning:
This indicates stable customer data. However, lower tenure customers may still represent a high-risk churn group.

## Bivariate Analysis 
Bivariate analysis is the process of studying the relationship between two variables at the same time. It helps us understand how one variable changes when another variable changes. In data analysis, it is used to find patterns, correlations, and comparisons between features. In this project, we use it to see how different customer attributes are related to churn.

1. Contract vs Churn 


```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Contract vs Churn")
plt.xticks(rotation=45)
plt.show()
```

### Observation:
Month-to-month contract customers show the highest churn compared to yearly or two-year contracts.

### Business Meaning:
Short-term contracts are less stable, and customers are more likely to leave. Encouraging long-term contracts can improve customer retention.

3. Correlation Heatmap

### Business Meaning:
These features are key drivers of churn. The company should focus on them for predictive modeling and retention strategies.
